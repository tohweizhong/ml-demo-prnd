```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import tensorflow as tf
import keras
from keras.utils import np_utils
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import SGD, Adadelta, Adagrad
```

```python
# import data
df = pd.read_csv("../data/train.csv")
# source: https://www.kaggle.com/mboaglio/simplifiedhuarus
# Classification of activities of daily living using smartphone inertial sensors

# first 5 rows
df.sample(5)
```

```python
# separate the target variable (y) with the features (X)
y = df["activity"]
X = df.drop(["activity"], axis = 1)
```

```python
# split into training and testing sets for cross-validation, 70:30
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                   test_size = 0.3,
                                                   random_state = 123)
```

```python
# encode y_train as integers instead of strings (required by Keras)
encoder = LabelEncoder()
encoder.fit(y_train)
y_train_encoded = encoder.transform(y_train)

# convert integers to dummy variables using one-hot encoding (required by Keras)
y_train_onehot = np_utils.to_categorical(y_train_encoded)

# repeat for y_test
encoder.fit(y_test)
y_test_encoded = encoder.transform(y_test)
y_test_onehot = np_utils.to_categorical(y_test_encoded)

print(y_train_onehot.shape)
print(y_test_onehot.shape)
```

```python
# build a simple sequential neural network model
n_cols = X.shape[1]
nn = Sequential()
nn.add(Dense(32, activation = "sigmoid", input_shape = (n_cols,)))
nn.add(Dropout(0.5))
nn.add(Dense(6, activation = "softmax"))

# compile
adadelta = Adagrad(lr=0.01, epsilon=None, decay=0.0)
nn.compile(optimizer= "adadelta", loss = "categorical_crossentropy", metrics = ["accuracy"])

# fit to training set using test set as validation
nn.fit(X_train, y_train_onehot, validation_data = (X_test, y_test_onehot), batch_size = 64, epochs = 2000)
```

```python
# view structure of model
nn.summary()
```

```python
# check model cross entropy loss and accuracy
score, acc = nn.evaluate(X_test, y_test_onehot)
print(score)
print(acc)
```
